```python
#Libraries import
import csv
from datetime import datetime
import requests
import pandas as pd
import matplotlib as plt
import matplotlib.pyplot as plt
```

```python
# Carga Archivo
df_covid = pd.read_csv('tmp/covid19_tweets.csv',encoding="utf-8")
```

```python
months_name = {
    1: 'January',
    2: 'Febrary',
    3: 'March',
    4: 'April',
    5: 'May',
    6: 'June',
    7: 'July',
    8: 'August',
    9: 'September',
    10: 'Octuber',
    11: 'December'
}

month_sort = [
    'January',
    'Febrary',
    'March',
    'April',
    'May',
    'June',
    'July',
    'August',
    'September',
    'Octuber',
    'December'
]
```

```python
#Limpiar data frame

#Convertimos locacion a string
casting_dict = {
    "user_name":'string',
    "user_location":'string',
    'user_created':'datetime64[ns]',
    'date':'datetime64[ns]'
}
df_covid_casted=df_covid.astype(casting_dict)

#filtrar valores mayores a 1990
df_covid_filter=df_covid_casted["user_created"] > "1990-01-01 00:00:00"
df_covid_casted=df_covid_casted[df_covid_filter]

#quitar ""
df_covid_casted['user_name'] = df_covid_casted['user_name'].str.strip()

#title
df_covid_casted['user_location'] = df_covid_casted['user_location'].str.title()

#Filtrar Data Frame valores nan
df_covid_filter=df_covid_casted["user_location"] != "Nan" 
df_covid_casted=df_covid_casted[df_covid_filter]

#agregamos columna de mes
df_covid_casted['month'] = df_covid_casted['user_created'].dt.month
df_covid_casted['week'] = df_covid_casted['user_created'].dt.week
df_covid_casted['hours'] = df_covid_casted['user_created'].dt.hour

#aplicamos nombre de mes
df_covid_casted['month']=df_covid_casted['month'].map(months_name)
```

```python
df_covid_casted
```

```python
#Funcion agrupa data frame 
def group_month():
    df_agrupado_month=df_covid_casted
    df_agrupado_month["count_month"]=df_agrupado_month["month"]
    df_agrupado_month=df_agrupado_month.groupby(["month"])['count_month'].count().reset_index(name='month_count')    
    return df_agrupado_month
```

```python
#Funcion agrupa data frame 
def group_weeks():
    df_agrupado_weeks=df_covid_casted
    df_agrupado_weeks["count_week"]=df_agrupado_weeks["week"]
    df_agrupado_weeks=df_agrupado_weeks.groupby(["month","week"])['count_week'].count().reset_index(name='count_week')    
    return df_agrupado_weeks
```

```python
#Funcion agrupa data frame 
def group_hours():
    df_agrupado_hours=df_covid_casted
    df_agrupado_hours["count_hour"]=df_agrupado_hours["month"]
    df_agrupado_hours=df_agrupado_hours.groupby(["hours"])['count_hour'].count().reset_index(name='count_hour') 
    df_agrupado_hours=df_agrupado_hours.sort_values(['count_hour'],ascending=False)
    return df_agrupado_hours
```

```python
#funcion grafica data frame
def grafica_time(df_agrupado_month,df_agrupado_weeks,df_agrupado_hours):
    fig,ax=plt.subplots(3,1)
    fig.set_size_inches(15,10)
    ax[0].bar(df_agrupado_month["month"],df_agrupado_month["month_count"])
    ax[2].scatter(df_agrupado_hours["hours"],df_agrupado_hours["count_hour"])
    #ax[2].bar(df_agrupado_weeks["week","month"],df_agrupado_weeks["count_week"])
    ax[1].bar(df_agrupado_weeks["week"],df_agrupado_weeks["count_week"])
    return plt.show()
```

```python
#Funcion agrupa data frame 
def group_mwh():
    df_agrupado_mwh=df_covid_casted
    df_agrupado_mwh["count_mwh"]=df_agrupado_mwh["month"]
    df_agrupado_mwh=df_agrupado_mwh.groupby(["month","week","hours"])['count_mwh'].count().reset_index(name='count_mwh')   
    df_agrupado_mwh=df_agrupado_mwh.sort_values(['month'],ascending=True)   
    return df_agrupado_mwh
```

```python
df_agrupado_month=group_month()
df_agrupado_weeks=group_weeks()
df_agrupado_hours=group_hours()
grafica_time(df_agrupado_month,df_agrupado_weeks,df_agrupado_hours)
df_agrupado_mwh=group_mwh()
```

```python
#Libraries import
import csv
from datetime import datetime
import requests
import pandas as pd
import matplotlib as plt
import matplotlib.pyplot as plt
```

```python
# Carga Archivo
df_covid = pd.read_csv('tmp/covid19_tweets.csv',encoding="utf-8")
```

```python
#Limpiar data frame

#Convertimos locacion a string
casting_dict = {
    "user_name":'string',
    "user_location":'string'
}
df_covid_casted=df_covid.astype(casting_dict)
#quitar ""
df_covid_casted['user_name'] = df_covid_casted['user_name'].str.strip()
#title
df_covid_casted['user_location'] = df_covid_casted['user_location'].str.title()
#Filtrar Data Frame valores nan
df_covid_filter=df_covid_casted["user_location"] != "Nan" 
df_covid_casted=df_covid_casted[df_covid_filter]
```

```python
#Funcion agrupa data frame 
def agrupado_usuarios_ciudad():
    df_agrupado=df_covid_casted
    df_agrupado["count"]=df_agrupado["user_name"]
    df_agrupado=df_agrupado.groupby(["user_location"])['count'].count().reset_index(name='user_count')
    return df_agrupado
```

```python
#Funcion ordena data frame
def ordenado_num_usuarios(df_agrupado_usuarios_ciudad):
    df_ordenado=df_agrupado_usuarios_ciudad.sort_values(['user_count'],ascending=False)
    return df_ordenado
```

```python
#funcion grafica data frame
def grafica_ubicaciÃ³n_geografica(df_ordenado):
    #Convertimos locacion a string
    casting_dict = {
    "user_location":'str'
    }
    df_ordenado=df_ordenado.astype(casting_dict)
    df_ordenado.plot(kind='bar',x='user_location',y='user_count')
    return plt.show()
```

```python
#Funcion filtra los mas reLevantes
def filtra_relevantes(df_ordenado):
    df_ordenado_filter=df_ordenado["user_count"]>500 
    df_ordenado_filter=df_ordenado[df_ordenado_filter]
    return df_ordenado_filter
```

```python
#Funcion filtra los mas reLevantes
def filtra_relevantes_mas(df_ordenado):
    df_ordenado_filter=df_ordenado["user_count"]>100 
    df_ordenado_filter=df_ordenado[df_ordenado_filter]
    return df_ordenado_filter
```

```python
#Funcion agrupa data frame 
def agrupado_usuarios_ciudad_mas():
    df_agrupado=df_covid_casted
    df_agrupado["count"]=df_agrupado["user_name"]
    df_agrupado=df_agrupado.groupby(["user_location","user_name"])['count'].count().reset_index(name='user_count')
    return df_agrupado
```

```python
df_agrupado_usuarios_ciudad=agrupado_usuarios_ciudad()
df_ordenado=ordenado_num_usuarios(df_agrupado_usuarios_ciudad)
df_ordenado_filtrado=filtra_relevantes(df_ordenado)
grafica_ubicaciÃ³n_geografica(df_ordenado_filtrado)
print(df_ordenado_filtrado)
```
